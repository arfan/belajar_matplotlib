```python
t = np.arange(0., 5., 0.2)
t
```

```python
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
         t, t**3, 'g^'
        )
plt.show()
```

```python
data = {
    'a': np.arange(50),
    'c': np.random.randint(0,50,50),
    'd': np.random.randn(50)
    
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
